```python
# Importar librerias standard de Qiskit
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit.circuit.library import *
from qiskit.visualization import *
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit_aer import *

import numpy as np

service = QiskitRuntimeService(channel="local")
```

```python
circuit = QuantumCircuit(3, 3)
space = Instruction(name=' ', num_qubits=0, num_clbits=0, params=[])

circuit.h(1)
circuit.cx(1, 2) #Creamos el entrelazamiento cu√°ntico entre los qubits 1 y 2.

alpha = np.random.random()
beta = np.sqrt(1 - alpha**2)
circuit.initialize([alpha, beta], 0) #Qubit a teletransportar con alpha y beta random

circuit.barrier()

circuit.cx(0, 1)
circuit.h(0)

circuit.barrier()

circuit.measure(0, 0)
circuit.measure(1, 1)

circuit.barrier()

with circuit.if_test((1, 1)):
    circuit.x(2)
with circuit.if_test((0, 1)):
    circuit.z(2)

circuit.save_statevector(label='final')

circuit.draw(output="mpl")
```

```python
simulator = Aer.get_backend('aer_simulator')

result = simulator.run(circuit).result()
statevector = result.data()['final']

display(array_to_latex(statevector))
```
